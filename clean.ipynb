```python
import pandas as pd
```

```python
df=pd.read_csv('train.csv')
td=pd.read_csv('test.csv')

print(df.head())
```

```python
age_table = df.groupby(["Pclass", "Sex"])["Age"].median()
age_table

test_table= td.groupby(["Pclass", "Sex"])["Age"].median()
test_table
```

```python
def fill_age_train(row):
    if pd.isnull(row["Age"]):
        return age_table.loc[row["Pclass"], row["Sex"]]
    return row["Age"]

def fill_age_test(row):
    if pd.isnull(row["Age"]):
        return test_table.loc[row["Pclass"], row["Sex"]]
    return row["Age"]

```

```python
df["Age"] = df.apply(fill_age_train, axis=1)
td["Age"] = td.apply(fill_age_test, axis=1)
```

```python
print(df["Age"].isnull().sum())
print(td["Age"].isnull().sum())

```

```python
df["Cabin"].fillna("Unknown", inplace=True)
td["Cabin"].fillna("Unknown", inplace=True)
```

```python
df["Embarked"].fillna("Unknown", inplace=True)
td["Embarked"].fillna("Unknown", inplace=True)
```

```python
df.to_csv('cleaned_train.csv', index=False)
td.to_csv('cleaned_test.csv', index=False)
```
